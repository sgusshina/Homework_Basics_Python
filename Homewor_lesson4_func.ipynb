```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def print_user():
    for document in documents:
        for directory, list_num in directories.items():
            if document['number'] in list_num:
                print(f"№ {document['number']}, тип: {document['type']}, владелец: {document['name']}, полка хранения: {directory}")


def print_directories():
    print(f"Текущий перечень полок: {', '.join(directories.keys())}")


def user_document():
    num_document = input('Введите номер документа: ')
    for document in documents:
        if document['number'] == num_document:
            print(f"Владелец документа: {document['name']}")
            return
    print('Документ не найден в базе')


def document_directories():
    num_document = input('Введите номер документа: ')
    for directory, list_num in directories.items():
        if num_document in list_num:
            print(f"Документ хранится на полке: {directory}")
            return
    print('Документ не найден в базе')


def full_information():
    print_user()


def append_directories():
    num_new_directory = input('Введите номер полки: ')
    if num_new_directory in directories:
        print("Такая полка уже существует.")
    else:
        directories[num_new_directory] = []
        print('Полка добавлена')
    print_directories()


def delete_directories():
    del_directory = input('Введите номер полки: ')
    if del_directory in directories:
        if not directories[del_directory]:
            del directories[del_directory]
            print("Полка удалена")
        else:
            print("На полке есть документы, удалите их перед удалением полки.")
    else:
        print('Такой полки нет')
    print_directories()


def append_document():
    num_doc = input('Введите номер документа: ')
    type_doc = input('Введите тип документа: ')
    user_ = input('Введите владельца документа: ')
    num_dir = input('Введите номер полки для хранения: ')
    if num_dir in directories:
        directories[num_dir].append(num_doc)
        documents.append({'type': type_doc, 'number': num_doc, 'name': user_})
        print('Документ добавлен.')
    else:
        print("Такой полки нет. Создайте её командой ads")
    print_user()


def del_documents():
    num_doc = input('Введите номер документа: ')
    for document in documents[:]:
        if document['number'] == num_doc:
            documents.remove(document)
            for dir_list in directories.values():
                if num_doc in dir_list:
                    dir_list.remove(num_doc)
            print('Документ удалён')
            print_user()
            return
    print('Документ не найден в базе')


def transfrom_doc():
    num_doc = input('Введите номер документа: ')
    num_new_dir = input('Введите номер полки: ')

    if num_new_dir not in directories:
        print('Такой полки не существует')
        print_directories()
        return

    for dir_list in directories.values():
        if num_doc in dir_list:
            dir_list.remove(num_doc)
            directories[num_new_dir].append(num_doc)
            print('Документ перемещён')
            print_user()
            return
    print('Документ не найден в базе')


def proggram():
    while True:
        command_user = input(
            'Введите команду: \n'
            'p - узнать владельца документа по его номеру;\n'
            's - узнать номер полки, на которой хранится документ;\n'
            'l - узнать полную информацию по всем документам, хранящимся в базе;\n'
            'ads - добавить новую полку;\n'
            'ds - удалить пустую полку;\n'
            'ad - добавить новый документ в данные;\n'
            'd - удалить документ из данных;\n'
            'm - переместить документ с полки на полку;\n'
            'q - выйти из программы.\n'
        )
        if command_user == 'q':
            print('Программа завершена.')
            break
        elif command_user == 'p':
            user_document()
        elif command_user == 's':
            document_directories()
        elif command_user == 'l':
            full_information()
        elif command_user == 'ads':
            append_directories()
        elif command_user == 'ds':
            delete_directories()
        elif command_user == 'ad':
            append_document()
        elif command_user == 'd':
            del_documents()
        elif command_user == 'm':
            transfrom_doc()
        else:
            print('Неверная команда. Попробуйте снова.')


proggram()


print(directories)
print(documents)



```
