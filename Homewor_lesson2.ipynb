```python
import math
import re


#Задание 1
string = 'testing'

print(string[int(len(string) / 2)], string[int(len(string) / 2 - 1)]) if len(string) % 2 == 0 else  print(string[int(len(string) / 2)])

# #Задание 2
list_num = []
num = int

while num !=0: 
    num = int(input('Введите число'))
    list_num.append(num)

print(sum(list_num))

# #Задание 3
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']

if len(boys) == len (girls): 
    dating = zip(boys,girls)
    print('Идеальные пары:')
    for couple in dating:
        print(f"{couple[0]} и {couple[1]}")
else: 
    print('Внимание, кто-то может остаться без пары')

# #Задание 4
countries_temperature = [
    ['Таиланд', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
    ['Германия', [57.2, 55.4, 59, 59, 53.6]],
    ['Россия', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
    ['Польша', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]
print ('Средняя температура в странах:')
for country in countries_temperature:
       temp_celsius = round(((sum(country[1])/len(country[1])) - 32) * 5/9, 1)
       print(f"{country[0]} - {temp_celsius}")

# # Задание 5
car_ids = ['А222ВС96', 'АБ22ВВ193']
pattern = '([АВЕКМНОР-УХ]{1})(\d{3})([АВЕКМНОР-УХ]{2})(\d{2,3})'
#  А, В, Е, К, М, Н, О, Р, С, Т, У, Х
for id in car_ids:
    check_id = re.fullmatch(pattern,id)
    if check_id:
        print(f"Номер {id} валиден. Регион: {check_id.group(4)}")
    else:
        print(f"Номер {id} не валиден")

# #Задание 6
stream = [
    'user4,2021-01-01;3',
    'user3,2022-01-07;4',
    'user2,2022-03-29;1',
    'user1,2020-04-04;13',
    'user2,2022-01-05;7',
    'user1,2021-06-14;4',
    'user3,2022-07-02;10',
    'user4,2021-03-21;19',
    'user4,2022-03-22;4',
    'user4,2022-04-22;8',
    'user4,2021-05-03;9',
    'user4,2022-05-11;11'
]
pattern_user = '(user\d{1}),(\d{4})-(\d{2})-(\d{2});(\d{1,4})'
count = 0
views_counter = 0
for user in stream:
    user_reg = re.fullmatch(pattern_user,user)
    if user_reg:
        count += 1
        views_counter += int(user_reg.group(5))

print(f"Среднее количество просмотров на уникального пользователя: {views_counter/count : .2f}")
        

```
