```python

##
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
command_user = input('Введите команду: \
                     p - узнать владельца документа по его номеру;\
                     s - узнать номер полки, на которой хранится документ;\
                     l - узнать полную информацию по всем документам, хранящимся в базе;\
                     ads - добавить новую полку;\
                     ds - удалить пустую полку;\
                     ad - добавить новый документ в данные;\
                     d - удалить документ из данных;\
                     m - переместить документ с полки на полку\
                     '
                     )
def print_user():
    for document in documents:
        for directorie,list_num in directories.items():
                if document['number'] in list_num:
                    print(f"№ {document['number']}, тип: {document['type']}, владелец: {document['name']}, полка хранения: {directorie}")

def print_directories(dir):
    print(f"Текущий перечень полок: {', '.join(list(dir.keys()))}")

def user_document():
    num_document = str(input('Введите номер документа'))
    doc_exist = False
    for document in documents:
        if document['number'] == num_document:
            doc_exist = True
        if doc_exist:
            print(f"Владелец документа: {document['name']}")
    if not doc_exist:
         print('Документ не найден в базе')


def document_directories():
    num_document = str(input('Введите номер документа'))
    doc_exist = False
    for directorie, list_num in directories.items():
        if num_document in list_num:
            doc_exist = True
            if doc_exist:
                print(f"Документ хранится на полке: {directorie}")
    if not doc_exist:
         print('Документ не найден в базе')

def full_information():
        print_user()
        

def append_directories():
    num_new_directorie = str(input('Введите номер полки'))
  
    if num_new_directorie in directories:
        print("Такая полка уже существует.")
    else:
        directories[num_new_directorie] = []
        print('Полка добавлена')
    print_directories(directories)
    return directories


def delete_directories():
    del_directories = str(input('Введите номер полки'))
    if del_directories in list(directories.keys()):
            if len(directories[del_directories]) == 0:
                directories.pop(del_directories)
                print("Полка удалена")
            else:
                print("На полке есть документы, удалите их перед удалением полки.")
    else:
            print('Такой полки нет')
    print_directories(directories)
    return directories

def append_document():
    num_doc = str(input('Введите номер документа'))
    type_doc = str(input('Введите тип документа'))
    user_ = str(input('Введите владельца документа'))
    num_dir = str(input('Введите номер полки для хранения'))
    if num_dir in list(directories.keys()):
        directories[num_dir].append(num_doc)
        new_user = {
            'type': type_doc, 'number': num_doc, 'name': user_
        }
        documents.append(new_user)
        print('Текущий список документов')
    else: 
        print("Такой полки нет. Создайте ее командой ads")
        
    print_user()
    return directories,documents

def del_documents():
    num_doc = str(input('Введите номер документа'))
    doc_exist = False
   
    for document in documents:
        if num_doc in document.values():
            documents.remove(document)
            doc_exist = True
    for dir,list_doc in directories.items():
        if num_doc in list_doc:
            list_doc.remove(num_doc)
    if doc_exist:
        print('Документ удален')
    else:
        print('Документ не найден в базе')
       
    print_user()
    return directories,documents

def transfrom_doc():
    num_doc = str(input('Введите номер документа: '))
    num_new_dir = str(input('Введите номер полки: '))
   
    document_exists = False
    
    for list_doc in directories.values():
        if num_doc in list_doc:
            document_exists = True
            print('Документ перемещен')
            break
    
    if not document_exists:
        print('Документ не найден в базе')
        
    if num_new_dir in directories:
        for dir, list_doc in directories.items():
            if num_doc in list_doc:
                list_doc.remove(num_doc)
                directories[num_new_dir] += [num_doc]
                break
    else:
        print('Такой полки не существует')
        print_directories(directories)
    
    print_user()

def main():
    global documents, directories
    while True:
        command_user = input(
            'Введите команду: \n'
            'p - узнать владельца документа по его номеру;\n'
            's - узнать номер полки, на которой хранится документ;\n'
            'l - узнать полную информацию по всем документам, хранящимся в базе;\n'
            'ads - добавить новую полку;\n'
            'ds - удалить пустую полку;\n'
            'ad - добавить новый документ в данные;\n'
            'd - удалить документ из данных;\n'
            'm - переместить документ с полки на полку;\n'
            'q - выйти из программы.\n'
        )
        if command_user == 'q':
            print('Программа завершена.')
            break
        elif command_user == 'p':
            user_document()
        elif command_user == 's':
            document_directories()
        elif command_user == 'l':
            full_information()
        elif command_user == 'ads':
            directories = append_directories()
        elif command_user == 'ds':
            directories = delete_directories()
        elif command_user == 'ad':
            documents, directories = append_document()
        elif command_user == 'd':
            documents, directories = del_documents()
        elif command_user == 'm':
            documents, directories = transfrom_doc()
        else:
            print('Неверная команда. Попробуйте снова.')
main()

print(directories)
print(documents)



```
